```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
import pandas as pd
from datasets import Dataset
nltk.download('punkt')
nltk.download('stopwords')
import numpy as np

from datasets import load_dataset, DatasetDict
from transformers import AutoTokenizer
```

```python
meanList = []
stdList = []
varianceList = []
```

# Seed 1

```python
def load_data(file_path):
    df = pd.read_csv(file_path)
    df['context'] = df['context'].fillna('').astype(str)
    return df

# Perform keyword extraction on each text
def perform_keyword_extraction(text_data):
    vectorizer = TfidfVectorizer(stop_words='english', tokenizer=word_tokenize, max_features=10, max_df=1.0)
    results = {}
    for index, text in enumerate(text_data):
        X = vectorizer.fit_transform([text])  # Fit TF-IDF on the individual text
        feature_names = vectorizer.get_feature_names_out()
        results[index] = feature_names
    return results
```

```python
# Main execution
file_path = "dataset/fine_tuning_dataset_seed_1.csv"
df = pd.read_csv(file_path)
df['context'] = df['context'].astype(str)
df['answer'] = df['answer'].astype(str)

dataset_dict = {
    "context": df["context"],
    "answer": df["answer"]
}

dataset = Dataset.from_dict(dataset_dict)

# Split the dataset into training and testing sets
splits = dataset.train_test_split(test_size=0.33, seed=1)

generated_ans_list = []
bigger_ans_list = []

# # Keyword Extraction
keywords_by_text = perform_keyword_extraction(splits['test']['context'])
for index, keywords in keywords_by_text.items():
    # print(f"Text {index+1} \n Keywords: {keywords}")
    for keyword in keywords:
        generated_ans_list.append(keyword)
    bigger_ans_list.append(generated_ans_list)
    generated_ans_list = []

print(bigger_ans_list)
```

```python
fullAnswers = list(splits['test']['answer'])
print(fullAnswers)

listofAnswers = []
for answer in fullAnswers:
    answer = answer.replace(',', ' ').split()
    listofAnswers.append(answer)
print(len(listofAnswers), "\n", listofAnswers)
```

```python
hitRateList = []
for i in range(len(bigger_ans_list)):
  Intersection = set(bigger_ans_list[i]).intersection(set(listofAnswers[i]))
  hitRate = round(len(Intersection) / len(listofAnswers[i]) * 100, 2)
  hitRateList.append(hitRate)

print(len(hitRateList), "\n", hitRateList)
```

```python
meanSeed1 = round(np.mean(hitRateList), 2)
stdSeed1 = round(np.std(hitRateList), 2)
varianceSeed1 = round(np.var(hitRateList), 2)



meanList.append(meanSeed1)
print("mean: ", meanList)
stdList.append(stdSeed1)
print("stdList: ", stdList)
varianceList.append(varianceSeed1)
print("varianceList: ", varianceList)
```
